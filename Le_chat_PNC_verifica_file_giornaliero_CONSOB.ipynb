```python
import pandas as pd

# ===============================================================================
# DaMISTRAL LE chat script di verifica PNC_verifica_file_giornaliero_CONSOB.ipynb
# ===============================================================================

# Carica il file Excel delle PNC
file_path = "PncPubbl.xlsx"
df = pd.read_excel(file_path)

# Seleziona solo le colonne utili: Emittente e PNC
pnc_df = df[['Emittente', 'Perc. posizione netta corta']].copy()

# Rinomina la colonna per chiarezza
pnc_df.columns = ['Emittente', 'PNC (%)']

# Pulizia: Converti la colonna in numerico, forzando gli errori a NaN
pnc_df['PNC (%)'] = pd.to_numeric(pnc_df['PNC (%)'], errors='coerce')

# Rimuovi eventuali righe con NaN (valori mancanti o non convertibili)
pnc_df = pnc_df.dropna(subset=['PNC (%)'])

# Somma le PNC per ogni emittente (nel caso ci siano più voci)
pnc_summary = pnc_df.groupby('Emittente')['PNC (%)'].sum().reset_index()

# Ordina per PNC decrescente
pnc_summary = pnc_summary.sort_values(by='PNC (%)', ascending=False)

# Arrotonda i valori a 2 decimali
pnc_summary['PNC (%)'] = pnc_summary['PNC (%)'].round(2)

# Salva il risultato in un nuovo file Excel
output_file = "PNC_Tutti_i_Titoli.xlsx"
pnc_summary.to_excel(output_file, index=False)

# Mostra il risultato a video
print("✅ Estrazione PNC completata. Ecco i risultati:")
print(pnc_summary.to_string(index=False))
```
